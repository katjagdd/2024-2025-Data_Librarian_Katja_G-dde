```python
import json
```

```python
url = "https://www.bibsonomy.org/json/search/Climate%20Change?items=1000&duplicates=merged"
```

```python
import urllib.request
```

```python
import urllib.parse
```

```python
f = urllib.request.urlopen(url) # f ist eine HTTPResponse
result = f.read().decode('utf-8') # utf-8 ist oft die richtige Wahl
```

```python
print(type(result))
```

```python
# Das Ergebnis ist ein String. Die Datei muss noch als JSON verarbeitet werden (Ergebnis = dict)
data = json.loads(result) # json.loads() lädt JSON Daten aus einem String
print(type(data))
```

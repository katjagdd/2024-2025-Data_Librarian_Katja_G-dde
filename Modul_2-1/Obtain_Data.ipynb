```python
import json
```

```python
url = "https://www.bibsonomy.org/json/search/Climate%20Change?items=1000&duplicates=merged"
```

```python
import urllib.request
```

```python
import urllib.parse
```

```python
f = urllib.request.urlopen(url) # f ist eine HTTPResponse
result = f.read().decode('utf-8') # utf-8 ist oft die richtige Wahl
```

```python
print(type(result))
```

```python
# Das Ergebnis ist ein String. Die Datei muss noch als JSON verarbeitet werden (Ergebnis = dict)
data = json.loads(result) # json.loads() lädt JSON Daten aus einem String
print(type(data))
```

```python
df_bibsonomy = pd.DataFrame(data["items"])
```

```python
df_bibsonomy
```

```python
df_bibsonomy["Search_Term"] = "Climate Change"
```

```python
df_bibsonomy
```

```python
df_bibsonomy_publications = df_bibsonomy.loc[df_bibsonomy.type == "Publication"]
df_bibsonomy_publications
```

```python
df_bibsonomy_publications.columns.to_list()
```

```python
df_bibsonomy_publications_selection = df_bibsonomy_publications[["type", "label", "journal", "year", "author"]]
df_bibsonomy_publications_selection
```

```python
df_bibsonomy_publications_selection = df_bibsonomy_publications_selection.rename(columns={"label": "title", "author": "authors"})
df_bibsonomy_publications_selection
```

```python
df_bibsonomy_publications_selection = df_bibsonomy_publications_selection.loc[df_bibsonomy_publications_selection["authors"].isnull() == False]
df_bibsonomy_publications_selection
```

```python
df_bibsonomy_publications_selection.loc[df_bibsonomy_publications_selection["title"].isnull() == False]
```

```python
df_bibsonomy_publications_selection.loc[df_bibsonomy_publications_selection["year"].isnull() == False]
```

```python
df_bibsonomy_publications_selection["year"].unique()
```

```python
df_bibsonomy_publications_selection["year_correct"] = df_bibsonomy_publications_selection.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_bibsonomy_publications_selection.loc[df_bibsonomy_publications_selection.year_correct == False]
```

```python
df_bibsonomy_publications_selection = df_bibsonomy_publications_selection.loc[df_bibsonomy_publications_selection.year_correct == True]
df_bibsonomy_publications_selection
```

```python
df_bibsonomy_publications_selection["year_number"] = df_bibsonomy_publications_selection["year"].astype(int)
df_bibsonomy_publications_selection
```
